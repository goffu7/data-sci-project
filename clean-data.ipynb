```python
import os
import glob
import json

# Path to the directory containing the JSON files
folder_path = "res\\2018"  # Update this to your folder path

# Use glob to get all JSON files in the directory
json_files = glob.glob(os.path.join(folder_path, "*"))

# Initialize a list to store the loaded JSON data
all_data = []

# Loop through each file and load the data
for file in json_files:
    with open(file, 'r', encoding='utf-8') as f:
        data = json.load(f)
        all_data.append(data)  # Append the loaded JSON to the list

# Check the number of files loaded and display one as an example
print(f"Loaded {len(all_data)} files.")
print(json.dumps(all_data[0], indent=2))  # Print the first JSON for inspection
```

```python
import pandas as pd

# Initialize an empty list to store the data
data_list = []

# Loop through each JSON object in `all_data`
for data in all_data:
    # Extract "status" information
    process_info = data.get("abstracts-retrieval-response", {}).get("authkeywords", {})

    if not isinstance(process_info, dict):
        continue
    
    # Extract authkeywords and the author-keyword list
    author_keywords = process_info.get("author-keyword", [])

    # year 
    year = '2018' 
    # Extracting citation-title
    bib = data.get("abstracts-retrieval-response", {}).get("item",{}).get("bibrecord", {})
    title = bib.get("head", {}).get("citation-title", None)


    subject_area = data.get("abstracts-retrieval-response", {}).get("subject-areas", {}).get("subject-area", [])
    if isinstance(subject_area, dict):
        continue 

    # Process 'subject-areas' array
    for subject in subject_area:
            # Loop through each keyword and append it to the data_list
        for keyword in author_keywords:
            # Extract the value of the keyword from "$"
            if isinstance(keyword, str):
                keyword_value = keyword[0]
            elif isinstance(keyword, dict):
                keyword_value = keyword.get("$", "")
            
            # Append the keyword value to the list
            data_list.append({"year":year,"title":title,"keyword": keyword_value,"subject_area": subject.get("$", ""),
            "subject_code": subject.get("@code", ""),
            "subject_abbrev": subject.get("@abbrev", ""),
            "fa": subject.get("@_fa", ""),})



    

# Convert the extracted data into a DataFrame
df = pd.DataFrame(data_list)

# Display the resulting DataFrame
print(df.drop_duplicates())
```
